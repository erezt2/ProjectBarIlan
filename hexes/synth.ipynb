```python
import os

# import required module
import os
# assign directory
directory = './'
def ends_or(file, *args):
    for x in args:
        if str(file).endswith(x):
            return True
    return False
# iterate over files in
# that directory
for filename in os.listdir(directory):
    # checking if it is a file
    if ends_or(filename, ".bin", ".eep", ".elf", ".sym",".lss",".map"):
        os.remove(filename)
    if filename.endswith("-CWLITEARM.hex"):
        try:
            os.remove(filename[:-14]+".hex")
        except FileNotFoundError:
            pass
        os.rename(filename, filename[:-14]+".hex")

```
